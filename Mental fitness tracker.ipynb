```python
df3 = df.merge(df2, on="Year")
```

```python
df3.head()
```

```python
df3.rename(columns={"Entity_x": "Entity"}, inplace=True)
df3.rename(columns={"Code_x": "Code"}, inplace=True)
```

```python
df3 = df3.drop(columns=['Entity_y'])
df3 = df3.drop(columns=['Code_y'])
df3 = df3.drop(columns=['Code'])
```

```python
df3.info()
```

```python
# Histogram of DALY
plt.hist(df3["daly"], bins=10)
plt.xlabel("DALY")
plt.ylabel("Frequency")
plt.title("Distribution of DALY")
plt.show()
```

```python
# Line plot of disorder percentages over the years
plt.figure(figsize=(20, 8))
plt.plot(df3["Year"], df3["schizophrenia"], label="Schizophrenia")
plt.plot(df3["Year"], df3["bipolar disorder"], label="Bipolar Disorder")
plt.plot(df3["Year"], df3["eating disorders"], label="Eating Disorders")
plt.plot(df3["Year"], df3["anxiety disorders"], label="Anxiety Disorders")
plt.plot(df3["Year"], df3["drug use disorders"], label="Drug Use Disorders")
plt.plot(df3["Year"], df3["depressive disorders"], label="Depressive Disorders")
plt.plot(df3["Year"], df3["alcohol use disorders"], label="Alcohol Use Disorders")

# Add more lines for other disorders
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.title("Disorder Percentages over the Years")
plt.legend()
plt.show()
```

```python
# Calculate the sum of disorder percentages
sum_percentages = df3[["schizophrenia", "bipolar disorder","eating disorders", "anxiety disorders","drug use disorders", "depressive disorders","alcohol use disorders"]].sum()

# Create a pie chart
plt.figure(figsize=(20, 6))
plt.pie(sum_percentages, labels=sum_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Disorder Percentages")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Bar plot of disorder percentages by country
plt.figure(figsize=(28, 8))
sns.barplot(data=df3, x="Entity", y="drug use disorders",)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Percentage")
plt.title("drug use disorders Percentage by Country")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Select the features and target variable
features = df3[["Year", "schizophrenia", "bipolar disorder","eating disorders", "anxiety disorders","drug use disorders", "depressive disorders","alcohol use disorders"]]
target = df3["daly"]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(" Linear Regression Mean Squared Error:", mse)
```

```python
from sklearn.cluster import KMeans

features_scaled = scaler.fit_transform(features)

# Perform K-means clustering
kmeans = KMeans(n_clusters=4, random_state=1234)
kmeans.fit(features_scaled)

# Add the cluster labels to the dataset
df3["Cluster"] = kmeans.labels_


# Visualize the clusters
plt.figure(figsize=(20, 6))
plt.scatter(df3["Year"], df3["daly"], c=df3["Cluster"], cmap="viridis")
plt.xlabel("Year")
plt.ylabel("schizophrenia percentage")
plt.title("K-means Clustering of Mental Fitness Tracker")
plt.show()
```
